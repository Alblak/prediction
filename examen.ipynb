```python
data["Sexe"]=data["Sexe"].map({
    'M':1, 'F':0
})
data["Maladies_chroniques"]=data["Maladies_chroniques"].map({
     'Aucune':0, 'Diabète':1, 'Asthme':2,'Hypertension':3
})

data["Allergies"]=data["Allergies"].map({
 
    'Aucune':0, 'Pénicilline':1, 'Aspirine':2, 'Latex':3, 'Pollen':4
   
})
data["Medicaments_actuels"]=data["Medicaments_actuels"].map({
   
     'Aucune':0,'Antihypertenseurs':1,'Metformine':2,'Inhalateur':3, 'Insuline':4
       
})
data["Type_chirurgie"]=data["Type_chirurgie"].map({
    'Appendicectomie': 1,
    'Prothèse de hanche': 2,
    'Cholécystectomie': 3,
    'Remplacement de genou': 4,
    'Hernie inguinale': 5,
    'Arthroscopie': 6,
    'Prostatectomie': 7,
    'Chirurgie de la colonne vertébrale': 8,
    'Chirurgie de la hernie': 9,
    "Remplacement de l'épaule": 10,
    'Chirurgie de la cheville': 11,
    'Chirurgie abdominale': 12,
    'Pontage coronarien': 13,
    'Chirurgie mammaire': 14,
    'Laparoscopie': 15,
    'Chirurgie du genou': 16,
    'Arthroplastie': 17,
    'Chirurgie de la hanche': 18,
    'Chirurgie de la main': 19,
    'Pontage gastrique': 20,
    'Remplacement de hanche': 21,
    'Chirurgie du coude': 22,
    'Laminectomie': 23,
    'Hystérectomie': 24,
    'Thyroïdectomie': 25
 
})
data["Anesthesie"]=data["Anesthesie"].map({
   
     'Générale':0,'Locale':1
       
})
data["Utilisation_de_produits_sanguins"]=data["Utilisation_de_produits_sanguins"].map({
   
     'Non':0,'Oui':1
       
})

data["Complications_post_operatoires"]=data["Complications_post_operatoires"].map({
   
     'Non':0,'Oui':1
       
})
```

```python
data
```

```python
data['Perte_sang'] = data['Perte_sang'].str.replace(' ml', '', regex=False)
```

```python
data=data.drop(['IMC','Duree_prevu_operation','Duree_reelle _opération', 'Complications_per_opératoires','Signes_vitaux','Douleur_post_operatoire','Temps_de_recuperation','Mobilisation précoce'], axis=1)
```

```python
data
```

```python
X=data.drop(['Complications_post_operatoires'], axis=1)
y=data['Complications_post_operatoires']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test=train_test_split(X,y, random_state=42, test_size=.2)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression


```

```python
modele_logistique=LogisticRegression()
```

```python
modele_logistique.fit(X_train, y_train)
```

```python
model_Random_Fores = RandomForestClassifier(n_estimators=100, random_state=42)

```

```python
model_Random_Fores.fit(X_train, y_train)
```

```python
Y_predictML=modele_logistique.predict(X_test)
Y_predictRF=model_Random_Fores.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

```python
classification_report(Y_predictML,y_test)
classification_report(Y_predictRF,y_test)
accuracy_score(Y_predictML,y_test)
accuracy_score(Y_predictRF,y_test)

confusion_matrix(Y_predictRF,y_test)
```
